# Dealing with 2D fuctions

```python
x = np.linspace(0,10,1000)
y = np.linspace(0,10,1000)
```

```python
xv,yv = np.meshgrid(x,y)
```

```python
x_ = np.linspace(-4,4,9)
y_ = np.linspace(-5,5,11)
xv_,yv_ = np.meshgrid(x_,y_)
print(xv_)
print("\n")
print(yv_)
```

```python
zv = xv_**2 + yv_**2
plt.contourf(xv_,yv_,zv,levels=30)
plt.colorbar()
```

# Basic Linear Algebra

```python
A = np.array([[3,2,1],[5,-5,5],[6,0,1]])
b1 = np.array([1,2,3])
b2 = np.array([3,4,5])
```

```python
print(A@b1) # perkalian matriks biasa
```

```python
A.T # Matriks Transpose
```

```python
print(np.dot(b1,b2))
```

```python
print(np.cross(b1,b2))
```

```python
a = np.array([[3,2,1],[5,-5,4],[6,0,1]])
c = np.array([4,3,0])
```

```python
np.linalg.solve(a,c)
```
