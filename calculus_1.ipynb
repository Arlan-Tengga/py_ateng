# Domain f(x,y)

```python
x = np.linspace(-2,2,400)
y = np.linspace(0,4,400)
X,Y = np.meshgrid(x,y)

# kondisi domain
domain = (Y >= X**2)
domain[(X==0)&(Y==1)] = False
plt.imshow(domain,extent=[-2,2,0,4],origin='lower',cmap="Blues",alpha=0.6)
plt.scatter(0,1,c="red",marker="x",s=80,label="excluded (0,1)")
plt.plot(x,x**2)
```

$f(x,y) = x^2y + \sqrt y$
